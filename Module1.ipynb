```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset with the correct delimiter
data = pd.read_csv('student_mat.csv', delimiter=';')
```

```python
# Display basic statistics
print(data.describe())

# Count unique values in categorical fields
print(data['sex'].value_counts())
print(data['Mjob'].value_counts())

# Visualize the distribution of grades
plt.figure(figsize=(10, 6))
sns.histplot(data['G3'], bins=10, kde=True)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)

# Drop rows with missing target values
data.dropna(subset=['G3'], inplace=True)

# Convert categorical fields to category type for optimization
data['sex'] = data['sex'].astype('category')
data['Mjob'] = data['Mjob'].astype('category')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='G3', data=data)
plt.title('Boxplot of Final Grades (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
sns.pairplot(data[['G1', 'G2', 'G3']], kind='scatter', diag_kind='kde')
plt.suptitle('Pair Plot of Grades (G1, G2, G3)', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_grades_by_mjob = data.groupby('Mjob')['G3'].mean().sort_values()
sns.barplot(x=avg_grades_by_mjob.index, y=avg_grades_by_mjob.values, palette='viridis')
plt.title('Average Final Grades (G3) by Mother\'s Job')
plt.xlabel('Mother\'s Job')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of final grades (G3)
plt.figure(figsize=(10, 6))
sns.histplot(data['G3'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot to compare final grades (G3) by study time
plt.figure(figsize=(10, 6))
sns.boxplot(x='studytime', y='G3', data=data, palette='Set2')
plt.title('Final Grades (G3) by Study Time')
plt.xlabel('Weekly Study Time')
plt.ylabel('Final Grade (G3)')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot of study time vs. final grades (G3)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='studytime', y='G3', data=data, color='purple')
plt.title('Study Time vs. Final Grades (G3)')
plt.xlabel('Weekly Study Time')
plt.ylabel('Final Grade (G3)')
plt.axhline(y=data['G3'].mean(), color='r', linestyle='--', label='Mean G3')
plt.legend()
plt.grid()
plt.show()
```

```python
# Bar plot of average final grades (G3) by parental education level
avg_grades_by_parental_education = data.groupby('famsize')['G3'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='famsize', y='G3', data=avg_grades_by_parental_education, palette='pastel')
plt.title('Average Final Grades (G3) by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Average Final Grade (G3)')
plt.grid(axis='y')
plt.show()
```

```python
# Count plot of final grades (G3) by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='G3', hue='sex', data=data, palette='muted')
plt.title('Distribution of Final Grades (G3) by Gender')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.legend(title='Gender')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data into a DataFrame
data = pd.read_csv('student_mat.csv')  # Uncomment and replace with your actual data loading code

# List of columns that contain 'yes'/'no' values
yes_no_columns = ['Dalc', 'Walc', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']

# Convert 'yes'/'no' columns to binary (1 and 0)
for column in yes_no_columns:
    if column in data.columns:
        data[column] = data[column].map({'yes': 1, 'no': 0})

# Convert other categorical columns into numeric using one-hot encoding
categorical_columns = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian']
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# Check for non-numeric columns
non_numeric_columns = data_encoded.select_dtypes(include=['object']).columns
if non_numeric_columns.any():
    print(f"Non-numeric columns found: {non_numeric_columns.tolist()}")

# Generate pair plot for numeric columns
plt.figure(figsize=(12, 10))  # Adjust the size of the figure if needed
pairplot = sns.pairplot(data_encoded, diag_kind='kde')
pairplot.fig.suptitle('Pairwise Relationships and Distributions', y=1.02)  # Title for the pair plot

# Save the pair plot as an image
plt.savefig('pairplot.png', bbox_inches='tight')  # Use bbox_inches to ensure the title fits in the saved image
plt.show()  # This will display the plot in the output cell
```
